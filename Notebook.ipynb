```python
y = data["Purchased Bike"]
X = data[["Income", "Age", "Education", "Home Owner"]]
```

```python
X
```

```python
y
```

```python
X
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
X["Education"] = le.fit_transform(X["Education"])
X["Home Owner"] = le.fit_transform(X["Home Owner"])
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.2)
```

```python
X_train.shape
```

```python
X.shape[0] * 0.8
```

```python
from sklearn.metrics import accuracy_score
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
param_grid = {
    "splitter" : ["best", "random"],
    "max-depth" : [None, 10, 20, 30, 40, 50],
    "min_sample_split": [2, 5, 10],
    "min_samples_leaf" : [1, 2, 4]
}
```

```python
tree_model = DecisionTreeClassifier()
```

```python
grid_search_tree = GridSearchCV(tree_model, param_grid)
```

```python
grid_search_tree.fit(X_train, y_train)
```

```python
grid_search_tree.best_params_
```

```python
grid_search_tree.predict(X_test)
```

```python
pred_tree = grid_search_tree.predict(X_test)
```

```python
accuracy_score(y_test, pred_tree)
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc_model = RandomForestClassifier()
```

```python
param_grid_rfc =  {
    "max_features" : [2, 3, 4],
    "n_estimators" : [32, 64, 128, 256]
}
```

```python
grid_rfc = GridSearchCV(rfc_model, param_grid_rfc)
```

```python
grid_rfc.fit(X_train, y_train)
```

```python
grid_rfc.best_params_
```

```python
pred_rfc = grid_rfc.predict(X_test)
```

```python
accuracy_score(y_test, pred_rfc)
```
